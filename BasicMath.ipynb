```python
# print all Prime numbers in an Interval
def is_prime(num):
  for i in range(2, num):
    if num%i == 0:
      return False
  return True

# Driver code
a, b = map(int, input("Enter start and end of range: ").split())

for i in range(a, b+1):
  if is_prime(i):
    print(i)
```

```python
# calculate factorial of a given number
def fact(num):
  fact = 1

  if num==0 or num==1:
    return 1

  for i in range(1, num+1):
    fact *= i

  return fact

#Driver Code
num = int(input("Enter a number: "))
factorial = fact(num)
print(f"Factorial of {num} is {factorial}")
```

```python
# sum of digits in a number
def sumOfNum(num):
    sum = 0
    while num>0:
      sum += num%10
      num //= 10
    return sum

# Driver Code
num = int(input("Enter a number: "))
sum_of_digits = sumOfNum(num)
print(f"Sum of digits of {num} is {sum_of_digits}")
```

```python
# n-th Fibonacci number
def fib_number(n):
  if n==0 or n==1:
    return n

  return fib_number(n-1) + fib_number(n-2)

# Driver Code
n = int(input("Enter which fibonacci element you want: "))
fib_num = fib_number(n)
print(f"The {n}th fibonacci element is {fib_num}")
```

```python
# count the number of digits in a number
def countDigits(num):
    cnt = 0
    while num>0:
      cnt += 1
      num //= 10
    return cnt

# Driver Code
num = int(input("Enter a number: "))
cntDigits = countDigits(num)
print(f"Number of digits in {num} is {cntDigits}")
```

```python
# Palindrome Number Check
def palNum(num):
  rev, temp = 0, num
  while temp>0:
    rev = (rev*10) + (temp%10)
    temp //= 10

  if rev==num:
     return True
  else:
    return False

# Driver Code
num = int(input("Enter a number: "))
palNumber = palNum(num)
if palNumber:
  print(f"The number {num} is a palindromic number.")
else:
  print(f"{num} is not a palindromic number.")
```
