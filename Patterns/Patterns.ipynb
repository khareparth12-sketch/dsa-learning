```python
def pattern1(n):
    for i in range(n):
        for j in range(n):
            print("*", end = "")
        print()

n = int(input("Enter number of rows: "))
pattern1(n)
```

```python
def pattern2(n):
    for i in range(n):
        for _ in range(i+1):
            print("*", end = "")
        print()

n = int(input("Enter number of rows: "))
pattern2(n)
```

```python
def pattern3(n):
    for i in range(1, n+1):
        for j in range(1, i+1):
            print(j, end="")
        print()

n = int(input("Enter number of rows: "))
pattern3(n)
```

```python
def pattern4(n):
    for i in range(1, n+1):
        for j in range(i):
            print(i, end="")
        print()

n = int(input("Enter number of rows: "))
pattern4(n)
```

```python
def pattern5(n):
    for i in range(n):
        for _ in range(n, i, -1):
            print("*", end="")
        print()

n = int(input("Enter number of rows: "))
pattern5(n)
```
