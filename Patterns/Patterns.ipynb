```python
def pattern1(n):
    for i in range(n):
        for j in range(n):
            print("*", end = "")
        print()

n = int(input("Enter number of rows: "))
pattern1(n)
```

```python
def pattern2(n):
    for i in range(n):
        for _ in range(i+1):
            print("*", end = "")
        print()

n = int(input("Enter number of rows: "))
pattern2(n)
```

```python
def pattern3(n):
    for i in range(1, n+1):
        for j in range(1, i+1):
            print(j, end="")
        print()

n = int(input("Enter number of rows: "))
pattern3(n)
```

```python
def pattern4(n):
    for i in range(1, n+1):
        for j in range(i):
            print(i, end="")
        print()

n = int(input("Enter number of rows: "))
pattern4(n)
```

```python
def pattern5(n):
    for i in range(n):
        for _ in range(n, i, -1):
            print("*", end="")
        print()

n = int(input("Enter number of rows: "))
pattern5(n)
```

```python
def pattern6(n):
    for i in range(n):
        for j in range(1, n-i+1):
            print(j, end=" ")
        print()

n = int(input("Enter number of rows: "))
pattern6(n)
```

```python
def pattern7(n):
    for i in range(n):
        # space
        for k in range(n-i-1):
            print(" ", end="")

        #star
        for _ in range(2*i+1):
            print("*", end="")

        #space
        for k in range(n-i-1):
            print(" ", end="")

        print()

n = int(input("Enter number of rows: "))
pattern7(n)
```

```python
def pattern8(n):
    for i in range(n):
        #space
        for k in range(i):
            print(" ", end="")

        #star
        for _ in range(2*n-(2*i+1)):
            print("*", end="")

        #space
        for k in range(i):
            print(" ", end="")
        print()

n = int(input("Enter number of rows: "))
pattern8(n)
```

```python
def pattern9(n):
    for i in range(n):
        # space
        for k in range(n-i-1):
            print(" ", end="")

        #star
        for _ in range(2*i+1):
            print("*", end="")

        #space
        for k in range(n-i-1):
            print(" ", end="")

        print()

    for i in range(n):
        #space
        for k in range(i):
            print(" ", end="")

        #star
        for _ in range(2*n-(2*i+1)):
            print("*", end="")

        #space
        for k in range(i):
            print(" ", end="")
        print()

n = int(input("Enter number of rows: "))
pattern9(n)
```

```python
def pattern10(n):
    for i in range(1, 2*n):
        stars = i
        if (i>n):
            stars = 2*n-i
        for _ in range(1, stars+1):
            print("*", end="")
        print()

n = int(input("Enter number of rows: "))
pattern10(n)
```
